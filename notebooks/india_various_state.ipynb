```python
import pandas as pd

df_new = pd.read_csv("../data/raw/india_varios.csv", low_memory=False)
```

```python
df_pm25 = df_new[df_new["pollutant_id"] == "PM2.5"]
df_pm25.shape
```

```python
df_pm25 = df_new[df_new["pollutant_id"] == "PM2.5"].copy()

df_pm25.shape
```

```python
df_pm25 = df_pm25.dropna(
    subset=["latitude", "longitude", "pollutant_avg", "state"]
)

df_pm25["pollutant_avg"] = pd.to_numeric(
    df_pm25["pollutant_avg"], errors="coerce"
)
```

```python
import plotly.express as px

fig = px.scatter_mapbox(
    df_pm25,
    lat="latitude",
    lon="longitude",
    color="pollutant_avg",
    size="pollutant_avg",
    hover_name="station",
    hover_data={
        "city": True,
        "state": True,
        "pollutant_avg": True
    },
    color_continuous_scale="RdYlGn_r",
    range_color=(10, 300),
    zoom=3.6,
    center={"lat": 22.5, "lon": 80},
    height=650,
    title="Near-Real-Time PM2.5 Levels Across Indian Monitoring Stations"
)

fig.update_layout(
    mapbox_style="carto-positron",
    margin={"r":0,"t":60,"l":0,"b":0},
    coloraxis_colorbar=dict(title="PM2.5 (g/m)")
)

fig.show()
fig.write_html(
    "../plots/all_india.html",
    include_plotlyjs="cdn",
    full_html=False
)
```

```python
df_pm25_clean = (
    df_new[df_new["pollutant_id"] == "PM2.5"]
    .dropna(subset=["latitude", "longitude", "pollutant_avg", "state"])
    .copy()
)

df_pm25_clean["pollutant_avg"] = pd.to_numeric(
    df_pm25_clean["pollutant_avg"], errors="coerce"
)

df_pm25_clean = df_pm25_clean.dropna(subset=["pollutant_avg"])
```

```python
df_pm25_clean = df_pm25_clean.rename(columns={
    "pollutant_avg": "pm25"
})
```

```python
df_pm25_clean.to_csv(
    "../data/processed/india_pm25_snapshot_clean.csv",
    index=False
)
```
